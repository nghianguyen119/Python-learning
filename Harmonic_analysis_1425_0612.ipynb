```python
import numpy as np
from scipy.fftpack import fft
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('py.csv', decimal=',', sep=';', usecols=[1,2])
title = pd.read_csv('py.csv', decimal=',', sep=';', usecols=[3])[0:2]
parameter_data = pd.read_csv('py.csv', decimal=',', sep=';', usecols=[4])[0:2]
```

```python
torque = data['Value']
```

```python
time = data['Time']
```

```python
plt.plot(time,torque)
```

```python
step = (time[1]-time[0])/np.power(10,3)
speed = parameter_data['Parameter'][0]
#number of point each period 
point_number = (1/step)/(speed/60)
#number of elimined period
num_eli = int(parameter_data['Parameter'][1])
```

```python
int(point_number)
```

```python
plt.plot(time[num_eli*int(point_number):],torque[num_eli*int(point_number):])
```

```python
plt.plot(time[10*int(point_number):11*int(point_number)],torque[10*int(point_number):11*int(point_number)])
```

```python
N = len(torque[int(point_number):])
# sample spacing
T = step

# Keep only 1/4 result
yf = fft(torque[int(point_number):])
xf = np.linspace(0.0, 1.0/(4.0*T), N/4)
yf[0] = yf[0]/2
plt.plot(xf, 2.0/N * np.abs(yf[0:N/4]),'-')
#plt.xlim(xmax = 4000)
title1 = title['Title'][0]
plt.title(title1)
plt.grid()
plt.show()
```
